Develop a program that would use appropriate hand coded morphological operations to enhance the given fingerprint.png image as the resultant output image.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
pic1=cv2.imread('fingerprint.png')
```

```python
kernel = np.ones((6,6),np.uint8)
erosion = cv2.erode(pic1,kernel,iterations = 1)
```

```python
plt.imshow(pic1)
```

```python
plt.imshow(erosion)
```

```python
dilation = cv2.dilate(erosion,kernel,iterations = 1)
plt.imshow(dilation)
```

```python
opening = cv2.morphologyEx(pic1, cv2.MORPH_OPEN, kernel)
plt.imshow(dilation)
```
